```python
import pandas as pd
```

```python
train_dataset = pd.read_csv('../dataset/train.csv', index_col=0)
test_dataset = pd.read_csv('../dataset/test.csv', index_col=0)
dataset = pd.concat([test_dataset, train_dataset])
dataset = dataset.drop(columns=['location', 'keyword', 'target'])
dataset.sort_index(inplace=True)
dataset
```

```python
df = pd.DataFrame()
df['has_insecure_link'] = dataset['text'].apply(lambda x: 1 if 'https://' in x else 0)
df['has_secure_link'] = dataset['text'].apply(lambda x: 1 if 'http://' in x else 0)
df['has_link'] = df['has_insecure_link'] | df['has_secure_link']
df['has_www'] = dataset['text'].apply(lambda x: 1 if 'www' in x else 0)
df['insecure_link_count'] = dataset['text'].apply(lambda x: x.count('http://'))
df['secure_link_count'] = dataset['text'].apply(lambda x: x.count('https://'))
df['link_count'] = df['insecure_link_count'] + df['secure_link_count']
```

```python
df['has_hashtagh'] = dataset['text'].apply(lambda x: 1 if '#' in x else 0)
df['hashtag_count'] = dataset['text'].apply(lambda x: x.count('#'))
```

```python
df['has_user_mentions'] = dataset['text'].apply(lambda x: 1 if '@' in x else 0)
df['user_mentions_count'] = dataset['text'].apply(lambda x: x.count('@'))
df
```
